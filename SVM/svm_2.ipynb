```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [2,2],
    [2,4],
    [6,2],
    [6,4]
])

y = np.array([-1,-1,1,1])

print(X)
print(y)
```

```python
plt.figure(figsize=(6,6))

plt.scatter(X[y==-1][:,0],
            X[y==-1][:,1],
            color='red',
            s=150,
            label='Class -1')

plt.scatter(X[y==1][:,0],
            X[y==1][:,1],
            color='blue',
            s=150,
            label='Class +1')

plt.grid(True)
plt.legend()

plt.show()
```

## Hyperplane Equation

The SVM decision boundary is represented by

w₁x₁ + w₂x₂ + b = 0

where

w = Weight vector

b = Bias

Our task is to find the correct values of w and b.

## Finding w and b

From the graph, the separating line is

x = 4

Can we convert this into the SVM equation?

w₁x + w₂y + b = 0

x = 4

x - 4 = 0 -> 1x + 0y -4 = 0

```python
w = np.array([1,0])

b = -4

print("Weight =",w)

print("Bias =",b)
```

## Prediction

Once we know w and b,

the SVM calculates

f(x)=w·x+b

If

f(x)>0

Predict Class +1

If

f(x)<0

Predict Class -1

```python
point=np.array([5,3])

score=np.dot(w,point)+b

print(score)
```

```python
point=np.array([3,3])

score=np.dot(w,point)+b

print(score)
```

## Functional Margin

The functional margin tells us whether every point is correctly classified.

Formula

γ = y(w·x+b)

If γ is positive,

the sample is correctly classified.

If γ is negative,

the sample is wrongly classified.

| Functional Margin | Meaning                                                      |
| ----------------: | ------------------------------------------------------------ |
|               > 1 | Correctly classified and outside the margin                  |
|               = 1 | Exactly on the margin (support vector)                       |
|            0 to 1 | Correctly classified but inside the margin (soft-margin SVM) |
|                 0 | On the decision boundary                                     |
|               < 0 | Misclassified                                                |


```python
margin=y*(np.dot(X,w)+b)

print(margin)
```

## Geometric Margin

Functional margin depends on the scale of w.

To obtain the real distance,

we divide by the length of w.

Geometric Margin

γ = Functional Margin / ||w||

```python
norm=np.linalg.norm(w) #distance from hyperplane to the closest point . sqrt(w1^2 + w2^2)

print(norm)
```

```python
geo_margin=(y*(np.dot(X,w)+b))/norm

print(geo_margin)
```

## Visualizing the Decision Boundary

The hyperplane is

w₁x+w₂y+b=0

Substituting

w=[1,0]

b=-4

gives

x=4

The margin lines are

w·x+b=1

and

w·x+b=-1

These are

x=5

and

x=3

```python
plt.figure(figsize=(6,6))

plt.scatter(X[y==-1][:,0],
            X[y==-1][:,1],
            color='red',
            s=150)

plt.scatter(X[y==1][:,0],
            X[y==1][:,1],
            color='blue',
            s=150)

plt.axvline(4,
            color='black',
            linewidth=2,
            label='Decision Boundary')

plt.axvline(3,
            color='green',
            linestyle='--',
            label='Margin')

plt.axvline(5,
            color='green',
            linestyle='--')

plt.xlim(0,8)

plt.ylim(0,6)

plt.grid(True)

plt.legend()

plt.show()
```

```python
A = np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
])

B = np.array([1, 1, 1])

w1, w2, b = np.linalg.solve(A, B)

# Step 4: Display the results
print("Weight Vector (w):")
print(f"w1 = {w1}")
print(f"w2 = {w2}")

print("\nBias:")
print(f"b = {b}")
```
