defintion of variance

```python

data = [2, 7, 3, 12, 9]
n = len(data)


mean = sum(data) / n


squared_diffs = [(x - mean) ** 2 for x in data]


variance = sum(squared_diffs) / (n - 1)


std_dev = variance ** 0.5

print(f"Variance: {variance:.2f}")  
print(f"Std Dev: {std_dev:.2f}")   
```

QR=Q3-Q1

LOWERLIMIT=QR*1.5+Q1
UPPER LIMIT =OR*1.5+Q3

```python
import pandas as pd
import numpy as np

df=pd.read_csv("Iris.csv")

q1=df["SepalLengthCm"].quantile(0.25)
q3=df["SepalLengthCm"].quantile(0.75)

qr=q3-q1

ll=q1-1.5*qr
ul=q3+1.5*qr

print("q1:",q1)
print("q3:",q3)
print("qr:",qr)
print("Lower Limit:",ll)
print("Upper Limit:",ul)
```

```python
o=df[(df["SepalLengthCm"]<ll) | (df["SepalLengthCm"]>ul)]
o
```

z-score 

z=(x-mue)/sigma->(value-mean)/Standard_deviation


\(\mathbf{z=}\frac{\mathbf{x-\mu }}{\mathbf{\sigma }}\)
Formula Breakdown
\(z\) = The Z-score (standard score)
\(x\) = The specific data point you want to score
\(\mu \) (mu) = The mean (average) of the entire dataset
\(\sigma \) (sigma) = The standard deviation of the entire dataset

```python
mean=df['Salary'].mean()
std=df['Salary'].std()

value=50000

z=(value-mean)/std

print(z)
```

```python
df['Salary_ZScore']=(
    df['Salary']-df['Salary'].mean()
)/df['Salary'].std(ddof=0)

df
```

```python
from sklearn.preprocessing import StandardScaler

s=StandardScaler()

df['Z-score']=s.fit_transform(df[['Salary']])

df
```
