```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
x = np.array([[1,1,5], [2,1,6], [3,2,5], [4,2,7], [5,3,8], [6,3,7], [7,4,9], [8,4,10]])
y = np.array([ 28, 35, 41, 50, 60, 63, 80, 90 ])
print(x)
print(y)
```

```python
weights=np.zeros(x.shape[1])

bias=0

print(weights)
```

```python
learning_rate = 0.001
epochs= 5000
n=len(x)
```

```python
cost_history=[]
bias_history=[]
weight_history=[]
```

```python

for epoch in range(epochs):
  dw=np.zeros(x.shape[1])
  db=0
  cost=0
  for i in range(n):
    prediction = bias
    for j in range(len(weights)):
      prediction+=weights[j]* x[i][j]
    error = prediction - y[i]
    cost += error**2
    for j in range(len(weights)):
      dw[j]=error * x[i][j]
    db+=error
  cost=cost/n
  dw=(2/n)*dw
  db=(2/n)*db

  weights = weights - learning_rate * dw
  bias = bias - learning_rate * db

  cost_history.append(cost)
  weight_history.append(weights.copy())
  bias_history.append(bias)

  if epoch % 500 == 0:
    print(f"Epoch {epoch}")
    print("Weights : ",weights)
    print("Bias : ",bias)
    print("Cost : ",cost)
    print(".................................")


```

```python
df_features = pd.DataFrame(x, columns=[f'Feature_{i}' for i in range(x.shape[1])])

g = sns.pairplot(df_features, diag_kind='kde', corner=True)
g.figure.suptitle('Pairwise Relationships of Continuous Columns', y=1.02)
plt.show()
```
