## Section 3: Conditional Statements

### Q11. Write a program that checks whether a given number (`num = -8`) is positive, negative, or zero.

```python
num=-8

if(num>0):
    print("Positive")
elif(num<0):
    print("Negative")
else:
    print("Zero")
```

### Q12. Write a program to check whether a given year (`year = 2024`) is a leap year.

```python
y=2024

if(y%400==0):
    print("Leap year")
else:
    if(y%100==0):
        print("Not leap year")
    else:
        if(y%4==0):
            print("Leap year")
        else:
            print("Not leap year")
```

### Q13. Write a grade calculator: given `marks = 78`, print the grade using the scale: ≥90 → A, ≥75 → B, ≥60 → C, ≥40 → D, below 40 → F.

```python
m=78

if(m>=90):
    print("A")
elif(m<90 and m>=75):
    print("B")
elif(m<75 and m>=60):
    print("C")
elif(m<60 and m>=40):
    print("D")
else:
    print("F")
```

### Q14. Write a program to check voting eligibility given `age = 17` (eligible if age is 18 or above).

```python
age=17

if(age>=18):
    print("Can vote")
else:
    print("Cannot vote")
```

### Q15. Using a **nested if**, check whether a given number (`num = 24`) is both even and positive.

```python
num=24
if(num>=0):
    if(num%2==0):
        print("Even and positive")
    else:
        print("odd and positive")
else:
    if(num%2==0):
        print("Even and negative")
    else:
        print("odd and negative")
```

## Section 5: `while` Loops

### Q20. Use a `while` loop to print numbers from 10 down to 1.

```python
n=10
while n>0:
    print(n)
    n=n-1
```

### Q21. Use a `while` loop to calculate the sum of the digits of the number `1234`.

```python
s=0
n=1234
while n>0:
    r=n%10
    s=s+r
    n=n//10
print("Sum:",s)
```

### Q22. Use a `while` loop with `break` to find the first number between 1 and 100 that is divisible by 7.

```python
n=1
while n<101:
    if(n%7==0):
        print(n)
        break;
    n=n+1
```

### Q23. Use a `while` loop with `continue` to print odd numbers from 1 to 10, skipping any number that is also a multiple of 3.

```python
n=1
while n<11:
    if(n%2!=0 and n%3==0):
        print(n)
    n=n+1
```
